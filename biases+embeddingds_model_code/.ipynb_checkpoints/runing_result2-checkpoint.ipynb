```python
latents_dimensions=[2**i for i in range(1, 6)]
best_test_rmses = []
for latent_dimension in latents_dimensions :

    model = AlternatingLeastSquare(data_dir, embedding_dim=latent_dimension)

    model.train_test_split()
    _, _, _, _, _, rmses_test= model.alternating_least_square(model.data_by_user_train, model.data_by_movie_train, lambd=0.1, tau=0.2, gamma=0.01, epochs = 5)
    best_test_rmses.append(min(rmses_test))
```

```python
import matplotlib.pyplot as plt
```

```python
_, ax= plt.subplots(figsize=(8,5))
ax.plot(latents_dimensions, best_test_rmses)
ax.set_xlabel("Latents dimensions", size = 18)
ax.set_ylabel("Test RMSEs", size = 18)
plt.savefig(f"plots/best_K.pdf", format="pdf", bbox_inches="tight", dpi=2000)
plt.show()
```

## finding the best lambda

```python
import numpy as np
```

```python
lambdas=np.round(np.linspace(0.0001,1,10), 1)
best_test_rmses = []
for lambda_ in lambdas[1:]:

    model = AlternatingLeastSquare(data_dir, embedding_dim=8)

    model.train_test_split()
    _, _, _, _, _, rmses_test= model.alternating_least_square(model.data_by_user_train, model.data_by_movie_train, lambd=lambda_, tau=0.2, gamma=0.01, epochs = 5)
    best_test_rmses.append(min(rmses_test))

```

```python
import matplotlib.pyplot as plt
```

```python
def plot_rmse(params, rmses,param_name, fig_name):
    _, ax= plt.subplots(figsize=(8,5))
    ax.plot(params, rmses)
    ax.set_xlabel(rf"$\{param_name}$", size = 18)
    ax.set_ylabel("Test RMSEs", size = 18)
    plt.savefig(f"plots/{fig_name}.pdf", format="pdf", bbox_inches="tight", dpi=2000)
    plt.show()
```

```python
plot_rmse(lambdas[1:], best_test_rmses, "lambda", "best_lambda")
```

## Optimal value for tau

```python
taus=np.round(np.linspace(0.00001,1,10), 1)
best_test_rmses = []
for tau in taus[1:]:

    model = AlternatingLeastSquare(data_dir, embedding_dim=8)

    model.train_test_split()
    _, _, _, _, _, rmses_test= model.alternating_least_square(model.data_by_user_train, model.data_by_movie_train, lambd=0.01, tau=tau, gamma=0.01, epochs = 5)
    best_test_rmses.append(min(rmses_test))
```

```python
plot_rmse(taus[1:], best_test_rmses, "tau", "best_tau")
```

## Optimal value for gamma

```python
gammas=np.round(np.linspace(0.00001,1,10), 2)
best_test_rmses = []
for gamma in gammas[1:]:

    model = AlternatingLeastSquare(data_dir, embedding_dim=8)

    model.train_test_split()
    _, _, _, _, _, rmses_test= model.alternating_least_square(model.data_by_user_train, model.data_by_movie_train, lambd=0.01, tau=0.2, gamma=gamma, epochs = 5)
    best_test_rmses.append(min(rmses_test))
```

```python
plot_rmse(gammas[1:], best_test_rmses, "gamma", "best_gamma")
```
