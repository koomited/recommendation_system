# Results on the 100K movilens datasets

```python
from alternating_least_square import AlternatingLeastSquare
```

# create object and get the data

```python
data_dir = '../data/ml-latest-small/ratings.csv'
als_100k_data = AlternatingLeastSquare(data_dir)
```

```python
als_100k_data.data_indexing()
```

## Plotting

### Power law

```python
als_100k_data.plot_power_law("power_law")
```

### Ratings distribution

```python
als_100k_data.plot_average_rating_hist()
```

## Train test split

```python
als_100k_data.train_test_split()
```

## Run the model: biases + embeddings

```python

user_biases, item_biases ,losses, rmse, losses_test, rmses_test= als_100k_data.alternating_least_square(als_100k_data.data_by_user_train, als_100k_data.data_by_movie_train, lambd=0.1, gamma=0.001, epochs = 20)
```

## Plot losses on training and test sets

```python
als_100k_data.line_plot(losses, losses_test, xaxis = "Iterations", yaxis = "Loss", fig_name="b_only_100k_train_test_lost")
```

## Plot RMSEs on training and test sets

```python
als_100k_data.line_plot(rmse, rmses_test, xaxis = "Iterations", yaxis = "RMSE", fig_name="b_only_100k_test_train_rmses")
```

## Training Losss only

```python
als_100k_data.plot_training_loss_only(losses,xaxis = "Iterations" , yaxis= "Loss", fig_name="b_only_100k_loss_train")
```
